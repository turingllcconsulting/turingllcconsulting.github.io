## Data Import and Cleanup

```python
# Importing library

import pandas as pd

# Reading csv file
meals = pd.read_csv('meals.csv')
meals = meals.drop(columns='UOM', inplace=False)

# Data preview
meals.head()
```

```python
# Cleaning the data

# Dropping duplicates
meals = meals.drop_duplicates(subset=['Sample', 'Test'])

# Pivoting data
meals = meals.pivot(index=['Sample', 'Sample_Date', 'Product', 'Plant'], columns='Test', values='Result').reset_index()
meals.fillna('', inplace=True)

# Setting data types
floatcol = ['Ash', 'Fat', 'Moisture', 'Peroxide', 'Protein']
meals[floatcol] = meals[floatcol].apply(pd.to_numeric, errors='coerce')

# Data preview
meals.head()
```

## Hypothesis Test

```python
# Importing library
from scipy.stats import spearmanr

# Defining class
class HypothesisTest(object):

    # Defining initialization function
    def __init__(self, data):
        self.data = data
        self.MakeModel()
        self.actual = self.TestStatistic(data)

    # Defining pvalue function with loop for 1000 iterations
    def PValue(self, iters=1000):
        self.test_stats = [self.TestStatistic(self.RunModel()) 
                           for _ in range(iters)]

        count = sum(1 for x in self.test_stats if x >= self.actual)
        return count / iters

    # Placeholder to calculate test statistic
    def TestStatistic(self, data):
        raise UnimplementedMethodException()

    # Placeholder to set up model
    def MakeModel(self):
        pass

    # Actually running the model
    def RunModel(self):
        raise UnimplementedMethodException()

# Defining class
class AshProteinTest(HypothesisTest):
    def __init__(self, data, ash_col, protein_col):
        self.data = data
        self.ash_col = ash_col
        self.protein_col = protein_col
        self.MakeModel()
        self.actual = self.TestStatistic(data)

    # Using Spearman for test statistic
    def TestStatistic(self, data):
        corr, _ = spearmanr(data[self.ash_col], data[self.protein_col])
        return corr

    # Null hypothesis with randomized permutations
    def RunModel(self):
        shuffled = self.data.copy()
        shuffled[self.protein_col] = np.random.permutation(shuffled[self.protein_col])
        return shuffled
```

```python
# Dropping NaNs
meals_cleaned = meals.dropna(subset=['Ash', 'Protein'])

# Running hypothesis test
ash_protein_test = AshProteinTest(meals_cleaned, 'Ash', 'Protein')
pvalue = ash_protein_test.PValue(iters=1000)

# Printing results
print(f"P-Value for Ash vs Protein Hypothesis Test: {pvalue:.3f}")
print(f"Actual Correlation: {ash_protein_test.actual:.3f}")

if pvalue < 0.05:
        print("Reject the null hypothesis: Ash is significantly correlated with Protein.")
else:
        print("Failed to reject the null hypothesis.")
```

## Regression Analysis

```python
# Importing libraries
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Dropping NaNs
meals_cleaned = meals.dropna(subset=['Ash', 'Protein'])

# Regression model
formula = 'Protein ~ Ash'
model = smf.ols(formula=formula, data=meals_cleaned).fit()

# Printing summary
print(model.summary())
```

```python
# Scatter plot of the observed data
plt.scatter(meals_cleaned['Ash'], meals_cleaned['Protein'], label='Data', alpha=0.7)

# Generate predictions using the model
x_values = np.linspace(meals_cleaned['Ash'].min(), meals_cleaned['Ash'].max(), 100)
y_values = model.predict(pd.DataFrame({'Ash': x_values}))

# Plot the regression line
plt.plot(x_values, y_values, color='red', label='Regression Line')

# Customize the plot
plt.xlabel('Ash')
plt.ylabel('Protein')
plt.title('Regression Analysis: Protein vs Ash')
plt.legend()
plt.grid(alpha=0.3)
plt.show()
```
